```python
book_text = gutenberg.raw('chesterton-brown.txt')
print(book_text[:1000])
```

```python
nltk.download('stopwords')
```

```python
from nltk.tokenize import RegexpTokenizer  #removing punctuation marks

tokenizer1 = RegexpTokenizer(r'\w+')
new_text = tokenizer1.tokenize(book_text)
print(new_text)
```

```python
from nltk.corpus import stopwords

```

```python
removed_stops = set(stopwords.words('english'))
```

```python
filtered_sentences = []
for w in new_text:
  if w not in removed_stops and w != 'the' and w!= 'The' and w!= 'THE':
    filtered_sentences.append(w)

print(filtered_sentences)
```

```python
from nltk.stem import WordNetLemmatizer #lemmatization
```

```python
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for w in filtered_sentences:
  print(w, " : " ,lemmatizer.lemmatize(w))
```

```python
from nltk import FreqDist #frequency of words
```

```python
split_result = [item.split() for item in filtered_sentences]
```

```python
flattened_list = [char for sublist in split_result for char in sublist]
```

```python
words = flattened_list
fdist1 = FreqDist(words)
print(fdist1)
print(fdist1.most_common())
```

```python
sent = " ".join(words)
```

```python
sent
```

```python
word_to_find = "equally"
word_occurrences = [word for word in filtered_sentences if word == word_to_find]
print(f"Occurrences of '{word_to_find}'-> {len(word_occurrences)}")

avg_chp_len = len(filtered_sentences)/len(word_occurrences)
print(f"Average chapter length is {avg_chp_len}")
```

```python
text_tags = nltk.pos_tag(filtered_sentences)                  #trigram
results = list()
for i, (txt, tag) in enumerate(text_tags):
    if tag in ["JJ", "JJR", "JJS"]:
        if (i > 0) and (i < len(text_tags)-1):
            results.append((text_tags[i-1][0], txt, text_tags[i+1][0]))
```
