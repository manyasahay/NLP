```python
import nltk
```

```python
nltk.download('gutenberg')
```

```python
#file identifiers in nltk Gutenberg corpus
corpus = nltk.corpus.gutenberg.fileids()
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
nltk.download('punkt')
```

```python
def analyze_texts(corpus):
    analysis_results = []

    for text in corpus:
        raw_text = text
        sentences = sent_tokenize(text)
        words = word_tokenize(text)

        num_sentences = len(sentences)
        num_words = len(words)
        avg_words_per_sentence = num_words / num_sentences if num_sentences > 0 else 0

        analysis_results.append({
            "raw_text": raw_text,
            "num_sentences": num_sentences,
            "num_words": num_words,
            "avg_words_per_sentence": avg_words_per_sentence
        })

    return analysis_results

analysis_results = analyze_texts(corpus)
for idx, analysis in enumerate(analysis_results):
    print(f"Text {idx + 1} Analysis:")
    print("Raw Text:", analysis["raw_text"])
    print("Number of Sentences:", analysis["num_sentences"])
    print("Number of Words:", analysis["num_words"])
    print("Average Words per Sentence:", analysis["avg_words_per_sentence"])
    print()
```

```python
from nltk.corpus import brown
```

```python
categories = brown.categories()
```

```python
print("Available categories in the Brown corpus:")
for category in categories:
    print(category)
```

```python
from nltk.probability import FreqDist
```

```python
def compare_categories(categories):
    for category in categories:
        print(f"Category: {category}")

        fileids = brown.fileids(category)

        total_words = 0
        total_sentences = 0
        word_freq_dist = FreqDist()

        for fileid in fileids:
            words = brown.words(fileid)
            sentences = brown.sents(fileid)
            total_words += len(words)
            total_sentences += len(sentences)
            word_freq_dist.update(words)

        print("Total Words:", total_words)
        print("Total Sentences:", total_sentences)

        print("\nTop 10 Most Common Words:")
        for word, freq in word_freq_dist.most_common(10):
            print(f"{word}: {freq}")


```

```python
# List of categories (genres) to compare
categories_to_compare = ['news', 'editorial', 'reviews', 'religion', 'hobbies']

compare_categories(categories_to_compare)
```

```python
from nltk.tokenize import word_tokenize
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
text = "This is a sample text for demonstrating tokenization, collocations, and finding words."
tokens = word_tokenize(text)
print("Tokens:", tokens)
finder = BigramCollocationFinder.from_words(tokens)
collocations = finder.nbest(BigramAssocMeasures().pmi, 10)
print("Collocations:", collocations)
target_word = "sample"
if target_word in tokens:
    print(f"'{target_word}' found in the text.")
else:
    print(f"'{target_word}' not found in the text.")
```
